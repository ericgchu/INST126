```python
import pandas as pd
def avg_minute_per_game(df):
    df['MPA'] = df['MP'] / df['G']
    mpg = df.groupby("Tm", as_index = False)["MPA"].mean()
    return mpg
def true_shooting(df):
    TS = df.groupby("Tm",as_index = False)["TS%"].mean()
    return TS
def three_point_attempt(df):
    three = df.groupby("Tm",as_index = False)["3PAr"].mean()
    return three
def get_team_avg(df):
    m_p_g = avg_minute_per_game(df)
    ts = true_shooting(df)
    par = three_point_attempt(df)
    print("\n*****AVG Team Stats*****\n")
    
    print("\nAVERAGE MINUTE PER GAME IS:\n")
    print(m_p_g)

    print("\nTRUE SHOOTING AVERAGE IS:\n")
    print(ts)

    print("\n3-POINT ATTEMPT RATE AVERAGE IS:\n")
    print(par)
dtypes={"MP":int, "TS%":float, "3PAr":float}
df = pd.read_csv("nba.csv",dtype = dtypes)
choice = int(input("Input a number"))
if choice == 4:
    get_team_avg(df)



    
    
```
